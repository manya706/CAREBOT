```python
import numpy as np
import pandas as pd
from sklearn import  preprocessing
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("Training.csv")
data
```

```python
def check_unique(data,columns):
    unique_values = {}
    for col in columns:
        unique_values[col] = data[col].unique().tolist()
    return unique_values

columns_list = list(data.columns)
print(check_unique(data,columns_list))
```

```python
data.isna().sum() #to check null values
```

```python
#Feature splitting
x = data.drop(columns=["prognosis"])
x.head()
x.shape
# data.shape
```

```python
y = data[['prognosis']]
y
```

```python
data_test = pd.read_csv("Testing.csv")
x_test = data_test.drop(columns=["prognosis"])
y_test = data_test[['prognosis']]
```

```python
y_test
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(y)
y = label.transform(y)
print(y)
print(y.shape)
label.fit(y_test)
y_test = label.transform(y_test)
print(y_test)
print(y_test.shape)
```

```python
y_train_set = set(y)
y_test_set = set(y_test)
if y_train_set == y_test_set:
    print("y_train and y_test encode the same values")
else:
    print("y_train and y_test do not encode the same values")
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(penalty = 'l2', C=0.1)
clf.fit(x,y)
y_pred_clf = clf.predict(x_test)
acc_clf = accuracy_score(y_test,y_pred_clf)
print(acc_clf)
```

```python
dt = DecisionTreeClassifier()
dt.fit(x,y)
y_predict_dt = dt.predict(x_test)
acc_dt = accuracy_score(y_test,y_predict_dt)
print(acc_dt)
```

TRYING TO FIX OVERFITTING USING:
1. K-FOLD CROSS VALIDATION

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(dt, x, y, cv=kfold)
print(scores)
```

```python
def random_forest(x,y,x_test,y_test):
    dt = RandomForestClassifier()
    dt.fit(x,y)
    y_pred = dt.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    return accuracy*100
    # print("Accuracy score by decision tree: ", accuracy_score)

a=random_forest(x , y, x_test, y_test)
print("Accuracy: ",a)
```

```python
from xgboost import XGBClassifier
def xgb_model():
    model = XGBClassifier()
    return model
xgb = xgb_model()
xgb.fit(x,y)
y_pred_xgb = xgb.predict(x_test)
accuracy = accuracy_score(y_test, y_pred_xgb)
print('Accuracy: {:.2f}%'.format(accuracy * 100))
```
